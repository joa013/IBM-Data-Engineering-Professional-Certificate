## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
df=pd.read_sql("select count(*) from CHICAGO_CRIME_DATA", engine) 
df
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
df=pd.read_sql("select community_area_name from CHICAGO_SOCIOECONOMIC_DATA where per_capita_income_ < 11000", engine) 
df 
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)



```python
df=pd.read_sql("select case_number from CHICAGO_CRIME_DATA where LOWER(DESCRIPTION) like '%minor%'", engine) 
df 
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
df=pd.read_sql("select * from CHICAGO_CRIME_DATA where lower(primary_type) = 'kidnapping' and lower(description) like '%child%'", engine) 
df 
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
df=pd.read_sql("select primary_type, description from chicago_crime_data where location_description like '%SCHOOL%'", engine) 
df 
```

### Problem 6

##### List the average safety score for each type of school.


```python
df=pd.read_sql("select elementary_middle_or_high_school, avg(safety_score) from chicago_public_schools group by elementary_middle_or_high_school", engine) 
df 
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line 


```python
df=pd.read_sql("select community_area_name,percent_households_below_poverty from chicago_socioeconomic_data order by percent_households_below_poverty desc fetch first 5 rows only", engine) 
df 
```

### Problem 8

##### Which community area is most crime prone?


```python
q = "select community_area_number, count(community_area_number) as crimes from chicago_crime_data group by community_area_number order by crimes desc limit 1"

df=pd.read_sql(q, engine) 
df
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
q = "select community_area_name from chicago_socioeconomic_data where hardship_index = (select max(hardship_index) from chicago_socioeconomic_data)"
df=pd.read_sql(q, engine) 
df
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
q= "select community_area_name from census_data where community_area_number in (select COMMUNITY_AREA_NUMBER from CHICAGO_CRIME_DATA GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(*) DESC LIMIT 1)"
df=pd.read_sql(q, engine) 
df
```

Queries to the database system catalog to retrieve table metadata

Retrieving the list of all tables in the schema:

```python
q="select * from syscat.tables where tabschema = 'RXN13110'"
df=pd.read_sql(q, engine) 
df
```

Count columns in the SCHOOLS table:

```python
q="select count(*) from syscat.columns where tabname = 'SCHOOLS'"
df=pd.read_sql(q, engine) 
df
```

Retrieve the the list of columns in SCHOOLS table and their column type (datatype) and length:

```python
q="select distinct(name), coltype, length from sysibm.syscolumns where tbname = 'SCHOOLS'"
df=pd.read_sql(q, engine) 
df
```

Additional queries

How many Elementary Schools are in the dataset?

```python
q="select count(*) from SCHOOLS where \"Elementary, Middle, or High School\" = 'ES'"
df=pd.read_sql(q, engine) 
df
```

What is the highest Safety Score?

```python
q="select max(SAFETY_SCORE) as max_safety_score from SCHOOLS"
df=pd.read_sql(q, engine) 
df
```

Which schools have highest Safety Score?

```python
q="select SCHOOL_ID,NAME_OF_SCHOOL from SCHOOLS where SAFETY_SCORE = (select MAX(Safety_Score) from SCHOOLS)"
df=pd.read_sql(q, engine) 
df
```

What are the top 10 schools with the highest "Average Student Attendance"?

```python
q="select school_id, name_of_school,AVERAGE_STUDENT_ATTENDANCE from SCHOOLS order by AVERAGE_STUDENT_ATTENDANCE desc nulls last limit 10"
df=pd.read_sql(q, engine) 
df
```

Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance

```python
q="select school_id, name_of_school,AVERAGE_STUDENT_ATTENDANCE from SCHOOLS \
order by AVERAGE_STUDENT_ATTENDANCE fetch first 5 rows only"
df=pd.read_sql(q, engine) 
df
```

Remove the '%' sign from the above result set for Average Student Attendance column

```python
q="select school_id, name_of_school,replace(AVERAGE_STUDENT_ATTENDANCE, '%', '') as avg_student_attndce from SCHOOLS \
order by AVERAGE_STUDENT_ATTENDANCE fetch first 5 rows only"
df=pd.read_sql(q, engine) 
df
```

Which Schools have Average Student Attendance lower than 70%?

```python
q="select school_id, name_of_school,AVERAGE_STUDENT_ATTENDANCE as avg_student_attndce \
from SCHOOLS \
where decimal(replace(AVERAGE_STUDENT_ATTENDANCE, '%', '')) < 70 \
order by Average_Student_Attendance"
df=pd.read_sql(q, engine) 
df
```

Get the total College Enrollment for each Community Area

```python
q="select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT \
   from SCHOOLS \
   group by Community_Area_Name"
df=pd.read_sql(q, engine) 
df
```

Get the 5 Community Areas with the least total College Enrollment sorted in ascending order

```python
q="select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT \
   from SCHOOLS \
   group by Community_Area_Name \
   order by TOTAL_ENROLLMENT \
   fetch first 5 rows only"
df=pd.read_sql(q, engine) 
df
```

List 5 schools with lowest safety score.

```python
q="select school_id, name_of_school, SAFETY_SCORE from SCHOOLS \
order by safety_score limit 5"
df=pd.read_sql(q, engine) 
df
```

Get the hardship index for the community area which has College Enrollment of 4368

```python
q="select hardship_index \
from CHICAGO_SOCIOECONOMIC_DATA CD, SCHOOLS CPS \
where CD.ca = CPS.community_area_number and College_Enrollment = 4368"
df=pd.read_sql(q, engine) 
df 
```

Get the hardship index for the community area which has the school with the highest enrollment.

```python
q="select ca, community_area_name, hardship_index \
from CHICAGO_SOCIOECONOMIC_DATA \
where ca in(select community_area_number from SCHOOLS order by college_enrollment desc limit 1)"
df= pd.read_sql(q, engine)
df 
```

Create a scatter plot using the variables per_capita_income_ and hardship_index. Explain the correlation between the two variables.

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
q = "SELECT per_capita_income_, hardship_index FROM chicago_socioeconomic_data"
df = pd.read_sql(q, engine)
plot = sns.jointplot(x='per_capita_income_',y='hardship_index', data=df)
```
